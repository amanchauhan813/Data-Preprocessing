```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['class_label','alcohol','malic_acid']
```

```python
df
```

```python
plt.title('Before MinMax Scaler')
sns.kdeplot(df.alcohol)
sns.kdeplot(df.malic_acid)
```

```python
df.iloc[:,0:1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:3],df.iloc[:,0:1],test_size=.3,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax=MinMaxScaler()
minmax.fit(x_train)
minmax_scaled_x_train=minmax.transform(x_train)
minmax_scaled_x_test=minmax.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(minmax_scaled_x_train,columns=x_train.columns)
x_test_scaled=pd.DataFrame(minmax_scaled_x_test,columns=x_test.columns)
```

```python
plt.title('After MinMax Scaler')
sns.kdeplot(x_train_scaled.alcohol)
sns.kdeplot(x_train_scaled.malic_acid)
```
